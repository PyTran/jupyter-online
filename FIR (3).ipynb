```python
# Lets make a filter
fs = 2000
coeffs = signal.firls(127, [0, 100, 200, 1000],[1,1, 0.0003, 0.0003],fs=fs)
plt.plot(coeffs)
freq,response = signal.freqz(coeffs)
plt.plot(0.5*fs*freq/np.pi, 20*np.log10(np.absolute(response)))
```
